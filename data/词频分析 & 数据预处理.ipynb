```python
import pandas as pd
import numpy as np
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
data=pd.read_excel("Problem_C_Data_Wordle.xlsx")
data
```

```python
mylist=np.array(data.iloc[0:1]).tolist()
data.columns = mylist 
data.drop([0], inplace=True)
```

```python
# 数据颠倒
data=data.iloc[::-1]
data
```

```python
# 比例
data["percentage"]=data.iloc[ : ,5]/data.iloc[:,4]
data
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# # 创建分类器
sid = SentimentIntensityAnalyzer()
# #英文情感分析
# for sentence in s2:
#     print(sentence)
#     print("情感得分", sid.polarity_scores(sentence))
senList=[]
attributes=data.iloc[:,3]
for i in attributes:
    senList.append(sid.polarity_scores(i))
compound=[]
for i in senList:
    compound.append(i["compound"])
compound
data["sentiment"]=compound
```

```python
from nltk.tag import pos_tag
a=np.array(pos_tag(attributes)).T[1]
data['attributes']=a
```

```python
with open("频率.txt", "r") as f:  # 打开文件
    txt = f.read()  # 读取文件
#     print(txt)
#     txt=txt.split("\n")
# txt=txt.splitlines()
# txt
```

```python
txt=txt.splitlines()
freq=[]
count=0
for i in range(len(txt)):
#     if (len(i)==0):
#         pass
#     else:
#         count=count+1
#         if (count%2==1):
#             txt[i]=txt[i].replace("{"," ")
#             txt[i]=txt[i].replace("}"," ")
#             freq.append(txt[i].split(","))
    if (i%2==1):
        txt[i]=txt[i].replace("{"," ")
        txt[i]=txt[i].replace("}"," ")
        freq.append(txt[i].split(","))
freq
newFreq=[]
for i in freq:
    for j in i:
        newFreq.append(j)
newFreq
# txt[0]
```

```python
import re
mathFreq=[]
for i in newFreq:
    a=re.split("\*|\^",i)
    if (len(a)==1):
        mathFreq.append(float(a[0]))
    else:
        mathFreq.append(float(a[0])*np.power(float(a[1]),float(a[2])))
#     print(type(a[]))
mathFreq
```

```python
wordZip=pd.DataFrame({'word':newList,'freq':mathFreq})
```

```python
wordZip.to_csv("wordZip.csv",sep=',',index=False,header=True)
```
