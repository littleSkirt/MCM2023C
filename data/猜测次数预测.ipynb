```python
def checkEnd(chosenWord,testWord):
    if(chosenWord==testWord):
        return True
    else:
        return False
```

```python
def probability(df):
    freq=np.array(df["freq"])
    prob=np.divide(freq,np.sum(freq))
    for i in range(1):
        index = np.random.choice(np.array(df["word"]), p=prob.ravel())
    return index
```
